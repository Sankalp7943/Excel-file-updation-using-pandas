```python
import numpy as np
import pandas as pd
```

```python
data_file_old=pd.read_excel('Desktop/Book_old.xlsx')
data_file_new=pd.read_excel('Desktop/Book_new.xlsx')

```

```python
arr1=data_file_old.values.tolist()
arr2=data_file_new.values.tolist()
ans=[]
```

```python
print(arr2[0][2])
```

```python
for i in arr1:
    if i[0] not in arr2:
        arr1.remove(i)
```

```python
for i in arr2:
    if i[0] not in arr1:
        ans.append(i)
    
```

```python
#print(ans)
```

```python
df = pd.DataFrame(ans) 
```

```python
print(df)
```

```python
writer = pd.ExcelWriter('pandas_simple.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
